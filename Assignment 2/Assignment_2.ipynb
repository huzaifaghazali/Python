```python
englishMarks = 75;
islMarks = 58;
mathsMarks = 90;
totalMarks = 300;
marksObtained = englishMarks + islMarks + mathsMarks
percentage = (marksObtained / totalMarks) * 100;

if percentage >= 90 and percentage < 100:
        print("Your grade is A+")
elif percentage >= 80 and percentage < 90:
        print("Your grade is A")
elif percentage >= 75 and percentage < 80:
        print("Your grade is B+")
elif percentage >= 70 and percentage < 75:
        print("Your grade is B")
elif percentage >= 65 and percentage < 70:
        print("Your grade is C+")
elif percentage >= 60 and percentage < 65:
        print("Your grade is C")
elif percentage >= 55 and percentage < 60:
        print("Your grade is D+")
elif percentage >= 50 and percentage < 55:
        print("Your grade is D")
else:
    print("Your grade is F")
    
print("Marks obtained",marksObtained,"out of 300:", )
print("Percentage : ", percentage)
        
```

```python
# 11.Write a Python program to get the largest number from a numeric list. 
def max_num(list):
    max = list [0]
    for a in list:
        if a > max:
            max = a
    return max
print(max_num([4, 7, -9, 10]))
```
